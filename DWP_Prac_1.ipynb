```python
import pandas as pd
import numpy as np
```

```python
# Sample datafram with missing values
data = {'Age': [20, 18, None, 32, None],
        'Gender': ['M', 'F', 'M', None, 'F']}
df = pd.DataFrame(data)
print(df)
```

```python
# Count of null values
print(df.isna().sum())
# Drop rows with missing values
df_dropped = df.dropna()
print(df_dropped)
```

```python
# Mean imputation
df['Age_Mean Imputed'] = df['Age'].fillna(df['Age'].mean())
print(df)

# Median imputation
df['Age_Median Imputed'] = df['Age'].fillna(df['Age'].median())
print(df)
```

```python
type(df['Age'])
```

```python
# Replace with minimum value
df['Age_Min Imputed'] = df['Age'].fillna(df['Age'].min())
print(df)

# Replace with maximum values
df['Age_Max Imputed'] = df['Age'].fillna(df['Age'].max())
print(df)
```

```python
# Arbitary value imputation
df['Age_Arbitary_Imputed'] = df['Age'].fillna(-1)
print(df)
```

```python
# Replace gender missing values with mode
df['Gender_freq_imputed'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df)
```
